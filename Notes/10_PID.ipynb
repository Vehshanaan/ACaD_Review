# 1. PID的形态

连续域的PID长这样：

$$
G(s) = K_p\left( 1+\frac{1}{T_is}+T_ds \right)
$$

其中：
 - $K_p$：比例增益
 - $T_i$：积分时间
 - $T_d$：微分时间

PID的设计就是摆弄这三个参数

# 2. 介绍数字PID

离散PID的表达式：

$$
G(z) = K_p\left( 1+\frac{Tz}{T_i(z-1)}+\frac{T_d(z-1)}{Tz} \right)
$$

其中，$T$是采样时间间隔
