# 3. 动态补偿器设计（频域方法）

设被控对象（plant）的传递函数是$G_p$，要设计的控制器传递函数是$G_c$:



目标闭环传递函数形式：



$$
G(s) = [I+G_o(s)]^{-1}G_o(s)
$$
其中，

$$
G_o(s) = G_p(s)G_c(s)
$$

从第一个式子，可以推得：

$$
G_o(s) = G(s)[I-G(s)]^{-1}
$$

$G_o$也是一个对角矩阵，因为$G(s)$和$I-G(s)$都是对角阵

**废话这么多，怎么算$G_c$呢？**

$$
G_c(s) = G_p^{-1}(s)G_o(s)
$$
