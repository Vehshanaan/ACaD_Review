# 1. 介绍

直接设计也叫离散设计、z平面设计

先把被控对象离散化，然后再在离散域中完成整个设计

主要是以下四招：
 - 根轨迹法
 - 频域响应法
 - RF法（讲这个）
 - PID法（调参，不是解析法）



# 2. RF法

对于待设计控制器$D$和被控对象$G$:



1. 确定闭环传递函数
   $$
   C(z) = \frac{D(z)G(z)}{1+D(z)G(z)}
   $$

2. 从上式计算控制器传函$D$:

 - 因果性原理：
   - $z\rightarrow\infty$处，C（闭环传函）的零点阶数要与G（被控对象）的零点阶数相同
 - 稳定性原理
   - $1-C$的零点必须包含所有$G$的单位圆外极点
   - $C$的零点必须包含所有$G$的单位圆外零点
 - 稳态精度：对于有速度信号追踪稳态误差$K_v$的Ⅰ型系统：
   - $C(z) |_{z=1}=1$
   - $-T\frac{dC(z)}{dz}|_{z=1}=\frac{1}{K_v}$

$$
D(z) = \frac{1}{G(z)}\frac{C(z)}{1-C(z)}
$$
