# 2. 能控性

**状态能控性**：如果一个动态系统是完全状态可控的，那在任何时间，可以构造一个不受约束的控制向量，此向量可以在有限时间内将任意状态转移到目标状态

验证方法：

$$P = [B, AB,  ... , A^{n-1}B]$$

其中，n是矩阵A的维度。

P被称为状态可控性矩阵。**如果rank(P) = n**，则系统完全状态可控

**输出能控性**：如果一个系统是完全是输出可控的，那对任意时间，可以构造一个不受约束的控制向量，此向量可以在有限时间内把任何输出转移到目标输出

验证方法：

$$Q = [CB, CAB, ..., CA^{n-1}B, D]$$

其中，n是矩阵A的维度。

Q被称为输出可控性矩阵，**如果rank(Q) = l(输出的维度)**，则系统完全输出可控

# 3. 能观性

定义：在时间$t$  ($t_0<t<T$)时，如果对所有的$t_0$和一部分$T$，每一个状态向量$x(t_0)$ 都可以用$t_0\sim T$的输出向量确定，那就称之为完全能观

从物理的角度来讲，就是说状态之间的转移一定会影响输出

验证方法：

$$
R = [C, CA, ..., CA^{n-1}]^T
$$

其中，n是矩阵A的维度。

**如果rank(R)=n**，就说系统是完全能观的
