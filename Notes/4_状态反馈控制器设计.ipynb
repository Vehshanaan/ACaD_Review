# 1. 状态反馈控制器

框图



维数描述：

 - u：m行1列
 - v: m行1列
 - y：l行1列
 - $\dot{x}$：n行1列
 - A: nxn
 - B: n*m
 - C: l*n
 - D: l*m

 - $F_s$: m*n，状态反馈增益矩阵，**用来调节闭环系统的极点**
 - $H$: m*m，输入前馈增益矩阵，**用来调节闭环系统的零点**

加上这两个控制器后的闭环传递函数表现为：



其中，

$$
\bar{F}_s = A+BF_s
$$

加上控制器之后，给被控对象的u就变成：
$$
u(t) = F_sx(t)+Hv(t)
$$

# 2. 如何设计$F_s$和$H$：

## 2.1 $F_s$:

根据闭环极点设计$F_s$

$$闭环传递函数的特征方程 = det[sI-\bar{F}_s]$$

$F_s$的大小：
 - 行数 = 输入的长度
 - 列数 = A的尺寸

## 2.2 $H$：

根据稳态误差设计$H$

对单位阶跃输入，H为：
$$
\begin{aligned}
H &= \left\{ [C+DF_s]\left[sI-\bar{F}_s\right]^{-1}B+D \right\}^{-1}_{s=0} \\
  &= -\left\{ [C+DF_s]\left[\bar{F}_s\right]^{-1}B-D     \right\}^{-1}
\end{aligned}
$$
