# 1. 状态空间模型表示

$$

\dot{x}(t) = Ax(t)+Bu(t)  (状态式)\\
y(t) = Cx(t)+Du(t) (输出式)

$$

其中各向量的维数：

**以下向量均为一列：**

x: n维

y: l维

u: m维

A (n*n)：描述系统的动态和状态向量x的轨迹

B (n*m)：展现输入u对状态x的影响

C (l*n)：将状态向量x转化为输出y

D (l*m)：表示输入u对输出y的直接影响

# 2. 状态方程转传递函数：

$$
\frac{Y(s)}{U(s)} = G(s) = C(sI_n-A)^{-1}B+D
$$

将系统传递函数转为状态方程被称为实现。

实现不是唯一的，即，一个传递函数可以有多组（A，B，C，D）的组合，它们都有同一个传递函数
