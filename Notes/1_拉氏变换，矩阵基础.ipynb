# 1. 拉氏变换

正变换公式：

$$
L[f(t)] = F(s) = \int\limits_0^\infty f(t)exp(-st)dt
$$

逆变换公式：

$$

L^{-1}[F(s)] = f(t) = \frac{1}{2\pi j}\int\limits_{c-j\infty}^{c+j\infty} F(s)exp(st)ds  (t > 0)

$$

为什么需要拉氏变换？
 - 经过拉氏变换后，可以将微分，积分等运算用复平面上的代数运算代替，再逆变换得到方程的解
 - 允许有效地利用图形方法预测系统性能，而不用解出微分方程。
 - 用拉氏变换解微分方程时可以同时得到解的瞬态和稳态分量（transient/steady-state component）
