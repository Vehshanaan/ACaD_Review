# 2. 计算流程



首先，观测器由以下三个组件构成：



他的组件就是$L,M,N$

对于$n$维的状态向量，$m$维的控制输入，$l$维的输出向量，这三个组件的维度：
 - $L$：n*n
 - $M$：n*m
 - $N$：n*l


1. **确定观测器的极点**

观测器必须要比原系统快，才能实时地反映系统的状态量变化

太快了会放大高频噪音（解释详见Week6 doc P3/5）

实践中，比原系统的极点偏左一丢丢就可以了

2. **计算N**

$$
极点特征式 = det[sI-(A-NC)]
$$
 

3. **计算L**

$$
L = A-NC
$$

4. **计算M**

$$
M = B-ND
$$
