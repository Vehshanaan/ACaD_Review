# 2. S域Z域极点位置与稳定性的关系

|s域|z域|稳定性|
|:-:|:-:|:-:|
|虚轴（$s=j\omega$）|单位圆（$\|z\|=1$）|恒定幅度震荡，不收敛|
|右半平面（$s$实部大于0）|单位圆外（$\|z\|> 0$）|发散|
|左半平面（$s$实部小于0）|单位圆内（$\|z\|< 0$）|收敛|

具体位置对应关系：





如果所有z传递函数的极点都位于单位圆内，则系统稳定。

单位圆外的极点只要有一个，系统就不稳定

位于单位圆上的成对极点会产生恒定震荡
